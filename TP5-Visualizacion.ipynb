<h1>Gráfico 3<h1>

<img src='https://www.consultantsmind.com/wp-content/uploads/2017/04/Consultantsmind-Airbus-Travel-Trips-by-Country.png'>

No existe escala en el eje Y.

El eje X no tiene valores, por lo que no se comprende correctamente lo que representa.

Muchos elementos, por lo que no se entienden los datos representados.

El tamaño de los círculos no es representativo en el gráfico.

Los elementos están muy desordenados, por lo que la comprensión se hace todavía más difícil.

Si bien se me hace muy complicado representar este gráfico de otra manera, la más acertada es a través de un Scatter Plot, ya que puedo representar los diferentes valores de cada elemento de manera clara e individual.


```python
ax = [2000,5000,7500,15000,16000,38000]
ay = [0,0.08,0.5,1,1.1,5]
bx = [5000,10000,16000,18000,30000]
by = [0.2,0.9,2,3,7]
cx = [8000,21000,26000]
cy = [0.5,2,3]
dx = [4000,5000,9000,10000]
dy = [0.09,0.2,0.5,0.8]
ex = [7000,9000,28000,3000,47000]
ey = [0.07,0.1,1,4,2]
fx = [2000,4000,11000]
fy = [0.2,0.3,0.7]
gx = [1000,3000,6000,8000]
gy = [0.01,0.04, 0.6,0.3]
plt.scatter(ax,ay,color='red', label='Asia Pacific') 
plt.scatter(bx,by,color='blue', label='Europe')
plt.scatter(cx,cy,color='black', label='North America')
plt.scatter(dx,dy,color='grey', label='Latin America')
plt.scatter(ex,ey,color='green', label='Middle East')
plt.scatter(fx,fy,color='yellow', label='CIS')
plt.scatter(gx,gy,color='brown', label='Africa')
plt.xlabel("GDP per Capita(Dollars)")
plt.ylabel("Trips per Capita")
plt.title("Propensity to Travel")
plt.legend(loc='best') 
plt.show()
```
